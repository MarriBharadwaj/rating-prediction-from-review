```python
import re
import string
import spacy
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nlp = spacy.load("en_core_web_sm")

def preprocess_text(text):
    if isinstance(text,str):
        # Lowercasing
        text = text.lower()
        # Handle contractions
        text = re.sub(r"n't", " not", text)
        text = re.sub(r"'s", " is", text)
        text = re.sub(r"'re", " are", text)
        text = re.sub(r"'d", " would", text)
        text = re.sub(r"'ll", " will", text)
        text = re.sub(r"'ve", " have", text)
        text = re.sub(r"'m", " am", text)
        # Remove special characters and formatting artifacts
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        # Tokenization
        tokens = word_tokenize(text)
        # Removing punctuation
        tokens = [token for token in tokens if token not in string.punctuation]
        # Removing stopwords
        stop_words = set(stopwords.words('english'))-{"aren't",'but','couldn',"couldn't",'didn',"didn't",'doesn',"doesn't", "don't", "hadn't", "hasn't", "haven't", 'if', "isn't", 'just', "mightn't",
        "mustn't", 'most', 'no', 'nor', 'not', 'only', "shouldn't", 'some', 'too', 'until', "wasn't", 'very', "won't", "wouldn't",}
        tokens = [token for token in tokens if token not in stop_words]
        # Lemmatization using SpaCy
        doc = nlp(" ".join(tokens))
        lemmatized_tokens = [token.lemma_ for token in doc]
        # Join tokens back into a string
        preprocessed_text = ' '.join(lemmatized_tokens)
        return preprocessed_text
    else:
        return ''

data['reviewText'] = data['reviewText'].apply(preprocess_text)
data['summary'] = data['summary'].apply(preprocess_text)
```

```python
from sklearn.model_selection import train_test_split

x = data[['reviewText', 'summary']]  # Features
y = data['overall']  # Target variable

x_train, x_temp, y_train, y_temp = train_test_split(x, y, test_size=0.3, random_state=42)

x_val, x_test, y_val, y_test = train_test_split(x_temp, y_temp, test_size=0.5, random_state=42)

print("Training set shape:", x_train.shape, y_train.shape)
print("Validation set shape:", x_val.shape, y_val.shape)
print("Test set shape:", x_test.shape, y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

combined_text_train = x_train['reviewText'] + ' ' + x_train['summary']  # Combine text from both columns
vectorizer = CountVectorizer()  # Initialize CountVectorizer
x_train_bow = vectorizer.fit_transform(combined_text_train)  # Fit and transform combined text

nb_classifier = MultinomialNB()  # Initialize Naive Bayes classifier
nb_classifier.fit(x_train_bow, y_train)  # Train the classifier using training data

combined_text_val = x_val['reviewText'] + ' ' + x_val['summary']  # Combine text from validation set
x_val_bow = vectorizer.transform(combined_text_val)  # Transform validation text data to bag-of-words
y_val_pred = nb_classifier.predict(x_val_bow)  # Predict ratings for validation data
val_accuracy = accuracy_score(y_val, y_val_pred)  # Calculate accuracy on validation data
print("Validation Accuracy:", val_accuracy)

combined_text_test = x_test['reviewText'] + ' ' + x_test['summary']  # Combine text from test set
x_test_bow = vectorizer.transform(combined_text_test)  # Transform test text data to bag-of-words
y_test_pred = nb_classifier.predict(x_test_bow)  # Predict ratings for test data
test_accuracy = accuracy_score(y_test, y_test_pred)  # Calculate accuracy on test data
print("Test Accuracy:", test_accuracy)
```

```python
from sklearn.naive_bayes import GaussianNB

gnb_classifier = GaussianNB()  # Initialize Gaussian Naive Bayes classifier
gnb_classifier.fit(x_train_bow.toarray(), y_train)  # Train the classifier using training data

y_val_pred_gnb = gnb_classifier.predict(x_val_bow.toarray())  # Predict ratings for validation data
val_accuracy_gnb = accuracy_score(y_val, y_val_pred_gnb)  # Calculate accuracy on validation data
print("Validation Accuracy (Gaussian Naive Bayes):", val_accuracy_gnb)

y_test_pred_gnb = gnb_classifier.predict(x_test_bow.toarray())  # Predict ratings for test data
test_accuracy_gnb = accuracy_score(y_test, y_test_pred_gnb)  # Calculate accuracy on test data
print("Test Accuracy (Gaussian Naive Bayes):", test_accuracy_gnb)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_classifier = DecisionTreeClassifier(criterion='entropy')  # Initialize Decision Tree classifier with entropy criterion
dt_classifier.fit(x_train_bow, y_train)  # Train the classifier using training data

y_val_pred_dt = dt_classifier.predict(x_val_bow)  # Predict ratings for validation data
val_accuracy_dt = accuracy_score(y_val, y_val_pred_dt)  # Calculate accuracy on validation data
print("Validation Accuracy (Decision Tree - Entropy):", val_accuracy_dt)

y_test_pred_dt = dt_classifier.predict(x_test_bow)  # Predict ratings for test data
test_accuracy_dt = accuracy_score(y_test, y_test_pred_dt)  # Calculate accuracy on test data
print("Test Accuracy (Decision Tree - Entropy):", test_accuracy_dt)
```
